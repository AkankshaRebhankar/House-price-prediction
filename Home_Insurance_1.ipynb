# Feature engineering

we extract month, day, or year, calculate the age of the client at the time of quotation, or determine the duration between the quotation date and the start of cover.

```python
# Convert date columns to datetime data type
df['QUOTE_DATE'] = pd.to_datetime(df['QUOTE_DATE'])
df['COVER_START'] = pd.to_datetime(df['COVER_START'])
df['P1_DOB'] = pd.to_datetime(df['P1_DOB'])

# Extract month, day, and year from QUOTE_DATE
df['QUOTE_MONTH'] = df['QUOTE_DATE'].dt.month
df['QUOTE_DAY'] = df['QUOTE_DATE'].dt.day
df['QUOTE_YEAR'] = df['QUOTE_DATE'].dt.year

# Calculate age of the client at the time of quotation
df['CLIENT_AGE_AT_QUOTE'] = (df['QUOTE_DATE'] - df['P1_DOB']).astype('<m8[Y]')

# Determine the duration between the quotation date and the start of cover
df['DURATION_QUOTE_TO_COVER'] = (df['COVER_START'] - df['QUOTE_DATE']).dt.days

# Display the modified dataframe with additional columns
print(df.head())
```

```python
# Select categorical columns for one-hot encoding
categorical_cols = ['P1_EMP_STATUS', 'P1_MAR_STATUS', 'PAYMENT_METHOD']

# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Display the modified dataframe with one-hot encoded columns
print(df_encoded.head())
```

```python
# Create a new column for total sum insured by adding SUM_INSURED_BUILDINGS and SUM_INSURED_CONTENTS
df['TOTAL_SUM_INSURED'] = df['SUM_INSURED_BUILDINGS'] + df['SUM_INSURED_CONTENTS']

# Display the DataFrame with the new column
print(df)
```
